```python
import pandas as pd
import matplotlib.pyplot as plt
orange = pd.read_csv('test.csv', sep=';', parse_dates=[['Date', 'Time']])
orange
```

```python
def plotZones(df):
    df = df.sort_values('Date_Time', ascending=True)
    plt.title('Zones')
    plt.ylabel('Minutes/Zone')
    plt.plot(df['Date_Time'], df['Grey'], color='grey')
    plt.plot(df['Date_Time'], df['Blue'], color='blue')
    plt.plot(df['Date_Time'], df['Green'], color='green')
    plt.plot(df['Date_Time'], df['Orange'], color='orange')
    plt.plot(df['Date_Time'], df['Red'], color='red')
    plt.xticks(rotation='vertical')
def plotCaloriesBurned(df):
    df = df.sort_values('Date_Time', ascending=True)
    plt.title('Calories Burned')
    plt.plot(df['Date_Time'], df['CALORIES BURNED'])
    plt.xticks(rotation='vertical')
def plotSplatPoints(df):
    df = df.sort_values('Date_Time', ascending=True)
    plt.title('Splat Points')
    plt.plot(df['Date_Time'], df['SPLAT POINTS'])
    plt.xticks(rotation='vertical')
def plotHeartRate(df):
    df = df.sort_values('Date_Time', ascending=True)
    plt.title('Heart Rate')
    plt.ylabel('BPM')
    plt.plot(df['Date_Time'], df['AVG. HEART-RATE'], color='blue')
    plt.plot(df['Date_Time'], df['Peak HR'], color='red')
    plt.xticks(rotation='vertical')
def plotTreadMillData(df):
    df = df.sort_values('Date_Time', ascending=True)
    plt.title('Tread Speed')
    plt.ylabel('MPH')
    plt.plot(df['Date_Time'], df['AVG. SPEED'], color='blue')
    plt.plot(df['Date_Time'], df['MAX SPEED'], color='red')
    plt.xticks(rotation='vertical')
```
